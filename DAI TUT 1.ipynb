No Null Values in the Data Set

Create the list of the string and numerical columns

```python
num_cols=data.select_dtypes(exclude=['object']).columns
num_cols
```

```python
str_cols=data.select_dtypes('object').columns
str_cols
```

```python
for col in num_cols:
    print("Column : ", col)
    print(f"Value counts of {col}: ",data[col].value_counts().sum())
    print(f"Number of Unique Values in {col}: ",data[col].nunique())
    print()
```

```python
for col in str_cols:
    print("Column : ", col)
    print(f"Value counts of {col}: ",data[col].value_counts().sum())
    print(f"Number of Unique Values in {col}: ",data[col].nunique())
    print()
```

```python
for col in num_cols:
    plt.figure()
    plt.boxplot(data[col])
    plt.title(col)
```

Pair Plot Visualization

```python
sns.pairplot(data)
```

Bar Plot Visualization


```python
plt.figure(figsize=(10,10))
plt.xticks(rotation=90)


sns.barplot(data, x = 'SYMBOL', y = '% CHANGE')
plt.show
```
